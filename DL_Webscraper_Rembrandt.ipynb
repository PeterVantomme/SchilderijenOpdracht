# Collectie van pages als resource
In onderstaande code kan je als uitkomst een array terugvinden met de links waar we de schilderijen van kunnen scrapen.

```python
from bs4 import BeautifulSoup, SoupStrainer
import requests
import re

html_home ="http://www.rembrandtpainting.net/complete_catalogue/"
html_page = requests.get("http://www.rembrandtpainting.net/complete_catalogue/complete_catalogue.htm")
soup_page1 = BeautifulSoup(html_page.content, 'html.parser')
table_url_pages = soup_page1.find('table', class_="tablelinks").find_all('a',href=True)
url_pages_collection = []

for link in table_url_pages:
    if link.has_attr('href'):
       url_pages_collection.append(html_home+str(link['href']))

#url_pages_collection
```

# Image URLs verzamelen
image_url_collection heeft als uitkomst een lijst met links naar de schilderijen op de site.

Deze kunnen we gebruiken om de images te downloaden.

```python
url_pages_content_collection = []

for link in url_pages_collection:
  url_pages_content_collection.append(BeautifulSoup(requests.get(link).content, "html.parser").find("table", class_="tablelinks").find_all("a",href=True))

image_url_collection = []

for i in range(len(url_pages_content_collection)):
  for link in url_pages_content_collection[i]:
    if link.has_attr('href'):
       image_url_collection.append(html_home+str(link['href']).replace("/SITES/essential%20vermeer/rembrandt", ""))

image_url_collection = list(set(image_url_collection))
```

# Afbeeldingen verkrijgen van website

```python
image_url_hi_res = []
good = bad = 0


for image_url in image_url_collection:
  try:
    image_url_collection_sample = BeautifulSoup(requests.get(image_url).content, "html.parser").find("div", {"id": "workimage"}).find("img").attrs.get("src")
    image_url_hi_res.append(image_url.rsplit('/', 1)[-2:][0]+"/"+image_url_collection_sample)
    good += 1
  except:
    bad +=1

print(str(good) +" images were processed, " + str(bad) + " images weren't processed." )
image_url_hi_res = list(set(image_url_hi_res))
```

# Afbeeldingen naar map

```python
import os
import shutil

count = 0
for image_url in image_url_hi_res:
  filename = str(count)+"Rembrandt"+".jpg"

  # Open the url image, set stream to True, this will return the stream content.
  r = requests.get(image_url, stream = True)

  # Check if the image was retrieved successfully
  if r.status_code == 200:  
    # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
    r.raw.decode_content = True
    
    # Open a local file with wb ( write binary ) permission.
    with open(filename,'wb') as f:
        shutil.copyfileobj(r.raw, f)
        
    print('Image sucessfully downloaded: ',filename)
  else:
    print('Image couldn\'t be retrieved')
  count +=1
```
